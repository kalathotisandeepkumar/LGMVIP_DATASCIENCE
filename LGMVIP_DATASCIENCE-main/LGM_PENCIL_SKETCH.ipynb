1.IMPORT THE LIBRARIES

```python
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

2.READ THE IMAGE

```python
def show(title,img,cmap='Reds'):
    plt.imshow(img,cmap=cmap)
    
```

```python
image=plt.imread(r"C:\Users\DELL\Downloads\tiger.jpg")
show("tiger",image,'Reds')
```

3.CONVERT ORIGINAL IMAGE TO GRAYSCALEIMAGE

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show("Gray Tiger", gray_image,'gray')
```

4.INVERT THE  GRAYSCALE IMAGE

```python
inverted_image = 255 - gray_image
show("inverted_tiger_image",inverted_image,'gray')
```

5.BLUR THE IMAGE BY USING THE GAUSSIAN FUNCTION

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
show("blurred_tiger_image",blurred,'gray')
```

6.PENCIL SKETCH IMAGE

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
show("pencil_sketch_tiger",pencil_sketch,'gray')
```
